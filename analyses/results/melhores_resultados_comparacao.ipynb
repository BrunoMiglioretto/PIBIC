```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
remove_datasets = [
    "DuckDuckGeese",
    "InsectWingbeat",
    "PEMS-SF",
]

df = pd.read_csv("../../results/results_07-07.csv")
df = df[~(df["dataset"].isin(remove_datasets))]

df_dtwd = pd.read_csv("../../metadata/uea_benchmark.csv")
```

```python
df
```

```python
best_config_df = df[
    (df["representation"] == "FIRTS")
    & (df["operation"] == "sum")
    & (df["convolution_algorithm"] == "Rocket")
]
best_config_df = pd.DataFrame(best_config_df[["dataset", "accuracy"]])
best_config_df = best_config_df.rename(columns={"accuracy": "firts_sum_rocket"})

rocket_df = df[
    (df["representation"].isnull())
    & (df["operation"].isnull())
    & (df["convolution_algorithm"] == "Rocket")
]
rocket_df = pd.DataFrame(rocket_df[["dataset", "accuracy"]])
rocket_df = rocket_df.rename(columns={"accuracy": "rocket"})

minirocket_df = df[
    (df["representation"].isnull())
    & (df["operation"].isnull())
    & (df["convolution_algorithm"] == "MiniRocket")
]
minirocket_df = pd.DataFrame(minirocket_df[["dataset", "accuracy"]])
minirocket_df = minirocket_df.rename(columns={"accuracy": "minirocket"})

unnormalised_dtwd_df = pd.DataFrame(df_dtwd[["dataset", "unnormalised_dtwd"]])
unnormalised_dtwd_df = unnormalised_dtwd_df.rename(columns={"unnormalised_dtwd": "DTWD"})

final_df = pd.merge(best_config_df, unnormalised_dtwd_df, on="dataset")
final_df = pd.merge(final_df, rocket_df, on="dataset", how="left")
final_df = pd.merge(final_df, minirocket_df, on="dataset", how="left")

final_df = final_df.drop_duplicates()

final_df
```
